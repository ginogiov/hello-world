```python
# Installiamo le librerie necessarie
!pip install yfinance yahooquery matplotlib

from yahooquery import search
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Inserisci qui il tuo ISIN
isin = "IE00BK5BQT80"

# 2. Cerchiamo il ticker Yahoo Finance
result = search(isin)

# Estraiamo il primo ticker trovato
try:
    ticker = result["quotes"][0]["symbol"]
    print("Ticker trovato:", ticker)
except:
    raise ValueError("Impossibile trovare un ticker per questo ISIN")

# 3. Scarichiamo i dati con yfinance
etf = yf.Ticker(ticker)
data = etf.history(period="2y")

print("\nPrime righe dei dati:")
print(data.head())

# 4. Grafico del prezzo di chiusura
plt.figure(figsize=(12,5))
plt.plot(data["Close"])
plt.title(f"Prezzo di chiusura - {ticker}")
plt.xlabel("Data")
plt.ylabel("Prezzo")
plt.grid(True)
plt.show()

# 5. Informazioni aggiuntive
print("\nInformazioni principali:")
info_keys = ["longName", "category", "fundFamily", "totalAssets", "navPrice", "yield", "expenseRatio"]
for key in info_keys:
    print(f"{key}: {etf.info.get(key)}")
```
